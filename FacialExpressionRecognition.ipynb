```python
video_path = "/content/drive/MyDrive/video_20231123_104107_edit.mp4"
import cv2
capture = cv2.VideoCapture(video_path)
face_model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
frame_list=[]
```

```python


while True:
    ret, frame = capture.read()

    if not ret:  # Check if the frame is valid
        break  # Exit the loop if there are no more frames

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_model.detectMultiScale(gray_frame, 1.1, 5)

    if len(faces) == 0:
        continue  # Skip this frame and go to the next one
    for (x, y, width, height) in faces:
       face_roi = gray_frame[y:y + height, x:x + width]

    if len(face_roi.shape) == 2:
        face_roi = cv2.cvtColor(face_roi, cv2.COLOR_GRAY2RGB)

    # Analyze the emotion on the face_roi
    emotion_list = DeepFace.analyze(face_roi, actions=["emotion"], enforce_detection=False)

    # Check if any face is detected in the ROI
    if len(emotion_list) > 0:
        dominant_emotion = emotion_list[0]['dominant_emotion']
        cv2.rectangle(frame, (x, y), (x + width, y + height), (255, 255, 0), 2)
        cv2.putText(frame, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_COMPLEX, 0.9, (255, 255, 0), 2)
    else:
        # Handle the case when no face is detected in the ROI
        print("No face detected in the ROI")

# Continue with the rest of your code...

    frame_list.append(frame)

capture.release()

# Check if at least one frame with a face was processed
if frame_list:
    height, width, colors = frame_list[0].shape
    size = (width, height)

    output_path = "Emotions.avi"
    output = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"DIVX"), 30, size)

    for frame in frame_list:
        output.write(frame)

    output.release()
    cv2.destroyAllWindows()
else:
    print("No frames with detected faces were processed.")
```
